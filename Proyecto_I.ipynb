## Ventas

```python
# Ver el número de ventas por año 
# Usando groupby
df.groupby("Year")["Global_Sales"].count()
```

```python
df.groupby("Year")["Global_Sales"].count().sort_values(ascending=False)
```

- Según este data el año con mas ventas fue el 2009, lo podemos observar en la siguiente gráfica

```python
# Total de ventas por año, plot
df.groupby("Year")["Global_Sales"].count().plot.bar()
```

```python
# Juegos más vendidos por año
df.groupby("Name")["Global_Sales"].count().sort_values(ascending=False)
```

```python
# Graficamos los 15 más vendidos
# Cambio a los 10 más vendidos 
df.groupby("Name")["Global_Sales"].mean().sort_values(ascending=False).head(10).plot.barh()
```
